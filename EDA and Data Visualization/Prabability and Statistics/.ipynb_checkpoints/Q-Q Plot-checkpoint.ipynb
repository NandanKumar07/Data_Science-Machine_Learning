### Q-Q plot

```python
import numpy as np
import pylab
import scipy.stats as stats
```

```python
# N(0,1)
std_normal = np.random.normal(loc = 0, scale = 1, size = 1000)
# here loc(location) is mean, and scale is standard deviation.
# 0 to 100th percentiles of std-normal
for i in range(0,101):
    print(i, np.percentile(std_normal,i))
```

```python
# generate random samples from N(20,5)

measurements = np.random.normal(loc = 20, scale = 5, size = 500)
# try size 1000 and 50
stats.probplot(measurements, dist = 'norm', plot = pylab)
pylab.show()
```

```python
# Lets try with uniform distribution

measurements = np.random.uniform(low = -1, high = 1, size = 10000)
# try size = 10000

stats.probplot(measurements, dist = "norm", plot = pylab)
pylab.show()
```
